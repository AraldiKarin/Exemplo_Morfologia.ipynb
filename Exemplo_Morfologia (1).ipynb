```python
#Esta célula esta importanto a biblioteca:
from google.colab.patches import cv2_imshow
```

```python
# Estas próximas 3 linhas estão carregando as imagens que serão apresentadas:
img = cv2.imread('j.png',0)
img_opening = cv2.imread('j_ruido.png',0)
img_closing = cv2.imread('j_furos.png',0)

# A linha seguinte irá obter as dimensões da imagem:
altura, largura = img.shape

# A linha seguinte irá obter o kernel:
kernel = np.ones((5,5),np.uint8)

# A linha seguinte irá apresentar o kernel:
print(kernel)
```

```python
# A linha seguinte irá aplicar a operação morfológica de erosão:
erosion = cv2.erode(img,kernel,iterations = 2)

# A linha seguinte irá aplicar a operação morfológica de dilatação:
dilation = cv2.dilate(img,kernel,iterations = 2)
```

```python
# A linha seguinte irá aplicar a operação morfológica de gradiente:
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)

# A linha seguinte irá aplicar a operação morfológica de abertura:
opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)

# A linha seguinte irá aplicar a operação morfológica de fechamento:
closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)
```

```python
# As próximas, e ultimas, linhas irão apresentar as imagens separadamente:
cv2_imshow(img)
cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
```
